```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# folder = 'experiment/'
def plot(folder):
    train_acc = np.load(folder +'/train_acc.npy')
    train_loss = np.load(folder+'/train_loss.npy')
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    plt.plot(np.arange(1, 1+train_acc.shape[0]), train_acc)
    plt.subplot(1,2,2)
    plt.plot(np.arange(1, 1+train_loss.shape[0]),train_loss)
    plt.show()
    try:
        val_acc = np.load(folder+'/val_acc.npy')
        val_loss = np.load(folder+'/val_loss.npy')
        plt.figure(figsize=(8,4))
        plt.subplot(1,2,1)
        plt.plot(3*np.arange(1, val_acc.shape[0]+1), val_acc,)
        plt.subplot(1,2,2)
        plt.plot(3*np.arange(1, val_acc.shape[0]+1), val_loss)
        print('Max val accuracy -> %f'%(np.max(val_acc)))
        print('Max train accuracy -> %f'%(np.max(train_acc)))
        print(val_acc)
    except:
        print('Val yet to run')
    return
```

```python
plot('trained_model/tiled_Ajit_0706')
```

```python
plot('trained_model/SAN_Ajit_0806/')
```

```python
plot('trained_model/Tiled_Ajit_0906/')
```

```python
def plot_curves(folders):
    plt.figure(figsize=(8,4))
    for folder in folders:
        train_acc = np.load(folder +'/val_acc.npy')
        train_loss = np.load(folder+'/val_loss.npy')
        plt.plot(3*np.arange(1, 11), train_acc[:10])
    
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend(['Tiled Attention Model','Stacked Attention Model'])
    plt.title("Validation Accuracy Curves for the two models")
        #plt.plot(np.arange(1, 1+train_loss.shape[0]),train_loss)
    plt.savefig('Curves/val_acc.png')
    plt.show()
    
```

```python
folders = ['trained_model/tiled_Ajit_0706','trained_model/SAN_Ajit_0806/']
plot_curves(folders)
```
