```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# folder = 'experiment/'
def plot(folder):
    train_acc = np.load(folder +'/train_acc.npy')
    train_loss = np.load(folder+'/train_loss.npy')
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    plt.plot(np.arange(1, 1+train_acc.shape[0]), train_acc)
    plt.subplot(1,2,2)
    plt.plot(np.arange(1, 1+train_loss.shape[0]),train_loss)
    plt.show()
    try:
        val_acc = np.load(folder+'/val_acc.npy')
        val_loss = np.load(folder+'/val_loss.npy')
        plt.figure(figsize=(8,4))
        plt.subplot(1,2,1)
        plt.plot(3*np.arange(1, val_acc.shape[0]+1), val_acc,)
        plt.subplot(1,2,2)
        plt.plot(3*np.arange(1, val_acc.shape[0]+1), val_loss)
        print('Max val accuracy -> %f'%(np.max(val_acc)))
        print(val_acc)
    except:
        print('Val yet to run')
    return
```

```python
plot('trained_model/tiled_0706/')
```
